```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url= "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
page=requests.get(url)
page    
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
table = soup.find("table", class_="wikitable sortable")
table
```

```python
for row in table.find_all("tr")[1:]:
    cells = row.find_all("td")
    rank = cells[0].text.strip()
    name = cells[0].text.strip()
    artist = cells[1].text.strip()
    upload_date = cells[3].text.strip()
    views = cells[2].text.strip()
    print(f"Rank: {rank}")
    print(f"Name:{name}")
    print(f"Artist: {artist}")
    print(f"Upload Date: {upload_date}")
    print(f"Views: {views}")
```

```python
url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"
page=requests.get(url)
page  
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
table = soup.find("table", class_="in-article sortable")
table
```

```python
novels = []
table = soup.find('table')
rows = table.find_all('tr')
for row in rows:
    columns = row.find_all('td')
    Title = columns[1].text.strip()
    Author= columns[2].text.strip()
    VolumesSales = columns[3].text.strip()
    Publisher = columns[4].text.strip()
    Genre = columns[5].text.strip()
    novel = {
        'Book Name': Title,
        'Author Name': Author,
        'Volumes Sold': VolumesSales,
        'Publisher': Publisher,
        'Genre': Genre
    }
    novels.append(novel)
    for novel in novels:
        print(novel)
```

```python
url = "https://www.imdb.com/list/ls095964455/"
page=requests.get(url)
page  
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
series_list = soup.find_all("div", class_="lister-item-content")
series_list
```

```python
for series in series_list:
    name = series.find("h3").find("a").text.strip()
    year_span = series.find("span", class_="lister-item-year").text.strip("()")
    genre = series.find("span", class_="genre").text.strip()
    runtime = series.find("span", class_="runtime").text.strip()
    rating = series.find("span", class_="ipl-rating-star__rating").text.strip()
    votes = series.find("span", attrs={"name": "nv"}).text.strip()
    print("Name:", name)
    print("Year Span:", year_span)
    print("Genre:", genre)
    print("Run Time:", runtime)
    print("Ratings:", rating)
    print("Votes:", votes)
    print()
```
